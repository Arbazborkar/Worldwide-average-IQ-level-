### Data visualization

```python
#countplot of IQ
plt.figure(figsize=(12,6))
sns.countplot(data=df,x="IQ",palette="rainbow")
plt.title("IQ countplot graph")
plt.xticks(rotation=45)
```

```python
top_10_country_count = df["country"].value_counts().nlargest(10)

plt.figure(figsize=(12,6))

sns.barplot(x=top_10_country_count.index, 
            y=top_10_country_count.values, 
            color="goldenrod")

plt.title("Top 10 Countries by Frequency")
plt.xlabel=("Country")
plt.ylabel=("Frequency")

plt.show()
```

```python
#DIstribution of estimates IQ
plt.figure(figsize=(12,6))
sns.histplot(df["IQ"],bins=20,kde=True)
plt.title("Distribution of estimated IQ")
plt.ylabel=("frequency")
plt.show()
```

```python
#Distribution of IQ score

plt.hist(data=df,x="IQ",color="orange",edgecolor="moccasin")
plt.title("Distribution of IQ level")
plt.xlabel=("IQ")
plt.ylabel=("count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data=df,x="avg_temp",y="IQ",color="green")
plt.title("Corelation between IQ an avg_temp")
```

```python
#distribution of average income

plt.figure(figsize=(12,6))
plt.hist(data=df,x="avg_income",color="pink",edgecolor="hotpink")
plt.title("Distibution of average income")
plt.show()
```

```python
#"Distribution of education expenditure"
plt.hist(data=df,x="education_expenditure",color="aquamarine",edgecolor="turquoise")
plt.title("Distribution of education expenditure")
plt.xlabel=("education_expenditure")
plt.ylabel=("count")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df,x="rank",y="education_expenditure")
plt.title("Relationship between rank and education expenditure")
plt.xticks(rotation=90)
plt.show()
```

```python
# education expenditure on IQ_

plt.figure(figsize=(10,6))
plt.scatter(data=df,y="IQ",x="education_expenditure",color="orangered")
plt.title("Education expenditure on IQ")
plt.show()
```

```python
#Correlation between Temperature and IQ

plt.figure(figsize=(10,6))
plt.scatter(data=df,x="avg_temp",y="IQ",color="darkviolet")
plt.title("Correlation between Temperature and IQ")
plt.show()
```
